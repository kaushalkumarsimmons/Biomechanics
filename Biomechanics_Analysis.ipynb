# Extract both subject Data

```python
sub1 = data_dict['Sub'][:,0]
sub2 = data_dict['Sub'][:,1]
```

# Extract the required dataset "STS_NCT_FS_SIT"

```python
# Here D1 denotes dataset for first subject and D2 denotes Dataset for second subject
D1=sub1[0]['STS_NCT_FS_SIT'][0][0]['L'][0][0]['hip_ang'][0][0]['y'][0][0]
D2=sub2[0]['STS_NCT_FS_SIT'][0][0]['L'][0][0]['hip_ang'][0][0]['y'][0][0]
```

Task. 2
# Plot for both subjects 

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("Plot for the STS_NCT_FS_SIT condition the average left hip angle in the y dimension.")
time_normalized=np.linspace(0,1001,1001)
plt.plot(time_normalized,D1)
plt.plot(time_normalized,D2)
plt.legend(['Sub1','.','.','Sub2'])
plt.xlabel('time')
plt.show()
```

In the above plot, upper three intersecting lines belongs to subject 1 and the lower three intersecting lines belongs to subject 2.
